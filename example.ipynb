```python
import matplotlib.pyplot as plt
import numpy as np
from main import *
```

```python
def get_graphs(fun):
    eps_s = [10 ** degree for degree in range(-20, 1)]
    points = [
        [(0, 1), (-5, 2)], 
        [(0, 10), (0, 0)],
        [(2, 1), (1, 0)]
    ]

    point_h, point_w = len(points), len(points[0])
    fig_x, axs_x = plt.subplots(point_h, point_w)
    fig_x.suptitle('By X')
    fig_y, axs_y = plt.subplots(point_h, point_w)
    fig_y.suptitle('By Y')

    for i in range(point_h):
        for j in range(point_w):
            p = points[i][j]
            # by x
            res_x = [np.abs(fun.df_dx(p) - df_dx(fun, p, eps)) for eps in eps_s]
            axs_x[i, j].plot(eps_s, res_x, '-ok')
            axs_x[i, j].set_xscale('log')
            axs_x[i, j].set_yscale('log')
            axs_x[i, j].set_title(f'Point: {p}')
            print(F"Res by X: {res_x}")
            # by y
            res_y = [np.abs(fun.df_dy(p) - df_dy(fun, p, eps)) for eps in eps_s]
            axs_y[i, j].plot(eps_s, res_y, '-ok')
            axs_y[i, j].set_xscale('log')
            axs_y[i, j].set_yscale('log')
            axs_y[i, j].set_title(f'Point: {p}')
            print(F"Res by Y: {res_y}")

    fig_x.tight_layout()
    fig_y.tight_layout()
```

#### Рассмотрим дважды дифференцируемую функцию в точке (0, 1)

$$f(x,y) = 3x^2 + xy + 2y^2 - x - 4 * y$$

```python
get_graphs(get_two_dif_fun())
```

Как видно из графиков, когда значения эпсилон слишком малы, то до определенного момента с ростом эпсилон возрастает точность вычисиления производной. Однако, начиная с некотого порога, во всех точках, с ростом эпсилон, начинает падать точность. В то же время, если рассмотреть точку локального минимума, то у неё точность расчета производной до некоторого порога близка к идеальной. 

При использовании метода конечных разностей для вычисления производных, слишком маленькие значения 
ε (эпсилон) могут привести к потере точности из-за ограниченной точности численных вычислений на компьютере.

Когда ε слишком маленькое, разность может стать очень близкой к нулю, и при вычитании таких близких значений из-за ограниченной точности чисел с плавающей запятой могут возникнуть проблемы округления. Это может привести к потере значащих цифр и, следовательно, к ухудшению точности вычисления производной.

Таким образом, хотя уменьшение ε в начале может увеличить точность, при слишком малых значениях 
ε может возникнуть потеря точности из-за ограниченной точности представления чисел в компьютере.

#### Рассмотрим однажды дифференцируемую функцию в точке (0, 0)

$$f(x,y) = (x + y)|x+y|$$

```python
get_graphs(get_one_dif_fun())
```

Тут наблюдается похожая картина. В начале из-за неточности вычислений при работе с слишком малыми значениями epsilon точность увеличивается, а затем уменьшается. Единственное отличие, что в точке глобального минимума точность падает с самого начале с ростом эпсилон.
